```python
def show_img(img):
    plt.imshow(img, cmap='gray')
    plt.axis('off')
    plt.show()
```

```python
img = cv.imread('sample.jpg', cv.IMREAD_UNCHANGED)

plt.imshow(img)
plt.axis('off')
plt.show()
```

```python
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
show_img(gray)
```

```python
blur = cv.GaussianBlur(gray, (15,1), cv.BORDER_REFLECT_101)
show_img(blur)
```

```python
blur = cv.GaussianBlur(gray, (1,15), cv.BORDER_REFLECT_101)
show_img(blur)
```

```python
blur = cv.GaussianBlur(gray, (15,15), cv.BORDER_REFLECT_101)
show_img(blur)
```

```python
edges = cv.Canny(blur, 50, 50, apertureSize=3)
show_img(edges)
```

```python
contours, _ = cv.findContours(edges, cv.RETR_TREE, cv.CHAIN_APPROX_SIMPLE)
img_contours = img.copy()
cv.drawContours(img_contours, contours, -1, (0,255,0), 2)

img_contours_rgb = cv.cvtColor(img_contours, cv.COLOR_BGR2RGB)
plt.imshow(img_contours_rgb)
plt.axis('off')
plt.show()
```

```python
img2 = cv.resize(img, (200, 200), interpolation=cv.BORDER_REFLECT)
show_img(img2)
```

```python
img2 = cv.resize(img, (200, 200))
show_img(img2)
```

```python
img = cv.imread("sample.jpg", cv.IMREAD_GRAYSCALE)
show_img(img)
```

```python
orb = cv.ORB.create()
```

```python
kp = orb.detect(img, None)
```

```python
kp, des = orb.compute(img, kp)
```

```python
img2 = cv.drawKeypoints(img, kp, None, color=(0,255,0), flags=0)
```

```python
plt.imshow(img2)
```
