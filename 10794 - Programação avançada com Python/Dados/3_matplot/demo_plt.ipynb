```python
x = np.array([1,2,3])
y = np.array([2,4,6])

y2 = np.sin(x)


plt.plot(x,y,
         color='red',
         linewidth=1,
         marker='o',
         markersize=5,
         markerfacecolor='#0f0f0f',
         markeredgecolor='y',
         linestyle='--',
         label="Dados",
         )


plt.plot(x,y2,
         color='y',
         linewidth=1,
         marker='o',
         markersize=5,
         markerfacecolor='#0f0f0f',
         markeredgecolor='y',
         linestyle='-',
         label="Dados",
         )



plt.xlabel("Eixo X")
plt.ylabel("Eixo Y")


plt.title("Dados", fontsize=25, fontweight="bold", loc="left", fontname="Courier" )


plt.legend()
plt.show()
```

```python
x = np.array([1,2,3])
y = np.array([2,4,6])

y2 = np.sin(x)


plt.figure(figsize=(10,5), dpi=100) # largura, altura

plt.plot(x,y,
         color='red',
         linewidth=1,
         marker='o',
         markersize=5,
         markerfacecolor='#0f0f0f',
         markeredgecolor='y',
         linestyle='--',
         label="Dados",
         )


plt.plot(x,y2,
         color='y',
         linewidth=1,
         marker='o',
         markersize=5,
         markerfacecolor='#0f0f0f',
         markeredgecolor='y',
         linestyle='-',
         label="Dados",
         )



plt.xlabel("Eixo X")
plt.ylabel("Eixo Y")


plt.title("Dados", fontsize=25, fontweight="bold", loc="left", fontname="Courier" )

plt.savefig("Dados_600.png", dpi=1200)



plt.legend()
plt.show()
```

```python
# Bar chart
lbl = ["Mas", "Peras", "Limes"]
val20 = [100, 231, 134]


b = plt.bar(lbl, val20, color="r", label="2020", alpha=0.5)
b[0].set_color("y")
b[2].set_color("b")
b[2].set_hatch("x")

b[1].set_hatch("*")
plt.legend()
plt.show()
```
